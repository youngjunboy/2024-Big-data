```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
mpg = pd.read_csv('mpg.csv')
```

```python
# 집단별 요약하기
# exam의 math 평균 구하기
# agg
exam.agg(mean_math = ('math','mean'))
```

### 집단별 요약 통계량 구하기

```python
# nclass별로 분리하기 .groupby()
# math 평균 .agg()

# 반별 수학 평균
exam.groupby('nclass').agg(mean_math = ('math','mean'))
```

```python
# 반별 영어 평균
exam.groupby('nclass').agg(mean_english = ('english','mean'))
```

```python
# 반별 과학 평균
exam.groupby('nclass').agg(mean_science = ('science','mean'))
```

```python
# 반별 과목 평균
exam.groupby('nclass').\
    agg(mean_science = ('science','mean'),
       mean_math = ('math','mean'),
       mean_english = ('english','mean'))
```

```python
# nclass 별로 분리 .groupby()
# 수학 점수 평균 mean
# 수학 점수 합계 sum
# 수학 점수 중앙값 median
# 빈도(학생수) count

exam.groupby('nclass')\
    .agg(mean_math = ('math','mean'),
        sum_math = ('math','sum'),
        median_math = ('math','median'),
        n = ('nclass','count'))
```

```python
# 모든 변수의 요약 통계량 한번에 구하기
# 과목별 평균
exam.groupby('nclass').mean()
```

```python
# 제조회사, 구동방식별로 구분
mpg.groupby(['manufacturer','drv']) \
        .agg(mean_cty = ('cty','mean'))
```

```python
### audi의 drv별 빈도
# audi 추출
# drv별 분리
# 빈도

mpg.query('manufacturer == "audi"') \
   .groupby('drv') \
   .agg(n = ('drv','count'))
```

```python
## chevrolet의 drv 별 빈도

# chevrolet 추출
# drv별 분리
# 빈도 구하기

mpg.query('manufacturer == "chevrolet"') \
   .groupby('drv') \
   .agg(n = ('drv','count'))
```

```python
# drv별 분리
# 빈도 구하기

mpg.groupby('drv') \
   .agg(n = ('drv','count'))
```

```python
# value_counts
mpg['drv'].value_counts()
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n으로 바꾸기 to_frame
# axis 이름에 drv 지정
# n이 100을 초과한 경우 추출

mpg['drv'].value_counts() \
          .to_frame('n') \
          .query('n > 100')
```

```python
# suv 추출 .query
# 합산 연비 변수 만들기 .assign total (hwy + cty) / 2
# 제조 회사별로 분리 .groupby
# 합산 연비 평균 구하기 .agg mean_tot
# 내림차순 정렬 .sort_values(assending = False)
# 1 ~ 5위까지 출력 .head()

mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total','mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head()
```

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id' : [1,2,3,4,5],
                      'midterm' : [60,80,70,90,85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id' : [1,2,3,4,5],
                      'final' : [70,93,65,95,80]})
```
