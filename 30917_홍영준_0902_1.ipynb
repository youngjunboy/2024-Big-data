## 데이터 전처리 - 원하는 형태로 데이터 가공하기

| 함수 | 기능
|-------|------|
| query() | 행 추출 |
| df[] | 열(변수) 추출 |
| sort_values() | 정렬 |
| groupby() | 집단별로 나누기 | 
| agg() | 통계치 구하기 |
| merge() | 데이터 합치기(열) |
| concat() | 데이터 합치기(행) |

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# exam에서 nclass가 3인 경우만 추출
exam.query('nclass==3')
```

```python
# exam에서 nclass가 3반이 아닌 경우만 추출
exam.query('nclass!=3')
```

```python
#  초과 미만 이상 이하
#  >   <   >=   <=
# 수학점수가 50점 초과한 경우 추출
exam.query('math > 50')
```

```python
# 여러 조건을 충족하는 행 추출하기
# 이거나 OR   |
# 이면서 AND   &
```

```python
# 1반이면서 수학점수가 50점 이상인 경우
# nclass == 1
# 이면서 &
# math >= 50
exam.query('nclass == 1 & math >= 50') 
```

```python
# 문제 모든 과목이 50점 이상인 경우 추출
# 모든 과목 And &
# math >= 50
# english >= 50
# science >= 50
exam.query('math >= 70 & english >= 70 & science >= 70')
```

```python
# 모든 과목 중에 한 과목이라도 90점이 넘는 과목의 경우 추출
# 이라도 OR
# math >= 90
# english >= 90
# science >= 90
exam.query('math >= 95 | english >= 95 | science >= 95')
```

```python
# 추출한 행으로 데이터 만들기

nclass1 = exam.query('nclass == 1')
```

```python
nclass1['math'].mean()
```

```python
# 원하는 열만 출력할 경우
exam[['nclass','math','english']].head()
```

```python
# 원하는 열을 제거하고 출력할 경우
# 메소드 .drop
# math 제거하고 출력
exam.drop(columns='math').head()
```

```python
# query + [] 조합 검색
# nclass가 1이고 영어만 출력
exam.query('nclass == 1')[['english']]
```

```python
# math 50점 이상인 행에서 id, math 추출
exam.query('math >= 50')[['id','math']].head(10)
```

```python
# 오름차순으로 정렬하기 .sort_valuse()
# 수학점수로 정렬
exam.sort_values('math').head()
```

```python
# 내림차순 속성 ascending = False
exam.sort_values('math', ascending = False).head()
```

```python
# nclass, math 오름차순
exam.sort_values(['nclass','math']).head(10)
```

```python
# nclass 내림차순 math 오름차순
exam.sort_values(['nclass','math'], ascending = [False, True]).head(10)
```

```python
# 변수 추가 .assign
# total
exam.assign(total = exam['math'] + exam['english'] + exam['science']).head(10)
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science'],     # total 추가
           mean = (exam['math'] + exam['english'] + exam['science']) / 3).head(10)  # mean 추가
```

### df.assign()에 np.where() 적용하기

```python
# test 열을 만들어주고 과학이 60점 이상이면 pass, fail
exam.assign(test = np.where(exam['science'] >= 60,'pass','fail'))
```
