```python
import pandas as pd
import numpy as np
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
mpg['total'].plot.hist()
```

```python
# 20 이상이면 pass, 그렇지 않으면 fail 부여 np.where = if
mpg['test'] = np.where(mpg['total'] >= 20,'pass','fail')
mpg.head()
```

```python
# 연비 합격 비도표 만들기
mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar(rot=0)
```

```python
# 문제 total 기준으로 A,B,C 등급 A 30 이상 B 20 이상 나머지는 C
# column name grade
mpg['grade'] = np.where(mpg['total'] >= 30,'A',
               np.where(mpg['total'] >= 20,'B','C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
# 문제 A,B,C,D 등급 만들기
# grade2 where 30 A 25 B 20 C 나머지는 D
mpg['grade2'] = np.where(mpg['total'] >= 30, 'A',
                np.where(mpg['total'] >= 25, 'B',
                np.where(mpg['total'] >= 20, 'C', 'D')))
mpg.head()
```

```python
count_grade2 = mpg['grade2'].value_counts().sort_index()
count_grade2
```

```python
count_grade2.plot.bar(rot=0)
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n 으로 바꾸기 to_frame
# axis 이름에 drv 지정
# n이 100을 초과한 경우 추출
mpg['drv'].value_counts() \
          .to_frame('n') \
          .query('n > 100')
```

```python
# suv 추출 .query
# 합산 연비 변수 만들기 .assign total (hwy + cty) / 2
# 제조 회사 별로 분리 .groupby
# 합산 연비 평균 구하기 .agg mean_tot
# 내림차순 정렬 .sort_values(ascending = False)
# 1 ~ 5위 까지 출력 .head()

mpg.query('category == "suv"').assign(total = (mpg['hwy'] + mpg['cty']) / 2).groupby('manufacturer').agg(mean_tot = ('total','mean')).sort_values('mean_tot',ascending = False).head()
```
