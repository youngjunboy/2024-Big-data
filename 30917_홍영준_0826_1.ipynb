```python
# mpg 데이터 불러오기
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg.tail()
```

```python
# mpg 통합연비 변수 만들기
# total = cty + hwy
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2 # 통합 연비 변수 만들기
mpg.head()
```

```python
# 20 이상이면 pass, 그렇지 않으면 fail 부여
# test >= 20 pass fail
mpg['test'] = np.where(mpg['total']>=20,'pass','tail')
mpg.head()
```
