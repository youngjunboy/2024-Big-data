```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# total 변수 추가
exam.assign(total = exam['math'] + exam['english'] + exam['science']).head()
```

```python
# mean 변수 추가
exam.assign(total = exam['math'] + exam['english'] + exam['science'], # total 추가
            mean = (exam['math'] + exam['english'] + exam['science']) / 3).head() # mean 추가
```

```python
# np.where 과학 점수가 60점 이상이면 pass, fail 출력
# test
exam.assign(test = np.where(exam['science'] >= 60,'pass','fail')).head()
```

```python
# 긴 데이터프레임명 지정
long_name = pd.read_csv('exam.csv')
long_name.head()
```

```python
# long_name에 new 변수 추가 : 수학 영어 과학 점수 합계
long_name.assign(new = exam['math'] + exam['english'] + exam['science']).head()
```

```python
# long_name 대신 x 입력
long_name.assign(new = lambda x: x['math'] + x['english'] + x['science']).head()
```

```python
# lambda 사용 mean 변수 추가
exam.assign(total = lambda x: x['math'] + x['english'] + x['science'], # total 추가
            mean = lambda x: x['total']/3).head() # mean 추가
```

```python
# 요약하기 agg
# math 평균 구하기
exam.agg(mean_math = ('math','mean'))
```

```python
# 클래스별 요약 통계량
# nclass별
# math 평균
exam.groupby('nclass') \
    .agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass', as_index = False).agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass').mean()
```
