```python
df = pd.DataFrame({'sex' : [1,2,1,8,2,1],
                  'score' : [5,4,3,4,2,6]})
df
```

```python
df['sex'].value_counts(sort=False).sort_index()
```

```python
df['score'].value_counts(sort=False).sort_index()
```

```python
# sex 8이면 NaN 할당
df['sex'] = np.where(df['sex'] == 8,np.nan,df['sex'])
df
```

```python
# score가 5보다 크면 NaN 할당
df['score'] = np.where(df['score']>5,np.nan,df['score'])
df
```

```python
# sex, score 결측치 제거 dropna
# sex별 분리 groupby
# score 평균 구하기 agg
df.dropna(subset = ['score','sex'])\
  .groupby('sex')\
  .agg(mean_score = ('score','mean'))
```
