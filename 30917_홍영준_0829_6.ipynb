## 06-1 데이터 전처리 - 원하는 형태로 데이터 가공하기
| 함수        | 기능         |
|-------------------|------------------|
| query() | 행 추출 |
| df[] | 열 (변수) 추출 |
| sort_values() | 정렬 |
| groupby() | 집단별로 나누기 |
| agg() | 통계치 구하기 |
| merge() | 데이터 합치기(열) |
| concat() | 데이터 합치기(행) |

```python
# 판다스 불러오기
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# 문제 수학 점수가 50점 초과한 경우만 출력
exam.query("math>50")
```

```python
# 문제 수학 점수가 50점 초과한 경우만 출력
exam.query("math<50")
```

```python
# 1반이면서 수학 점가 50점 이상인 경우 추출
exam.query("nclass == 1 & math >= 50")
```

```python
# 2반이면서 영어 점수가 80점 이상인 경우
exam.query("nclass == 2 & english >= 80")
```

```python
# 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query("math >= 90 | english >= 90")
```

```python
# 영어점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query("english < 90 | science < 50")
```

```python
# 1,3,5반 해당하는 값 출력
# exam.query("nclass == 1 | nclass == 3 | nclass == 5")
exam.query("nclass in [1,3,5]")
```

```python
# nclass가 1인 행 추출해서 nclass1 할당
nclass1 = exam.query("nclass == 1")
# nclass가 2인 행 추출해서 nclass1 할당
nclass2 = exam.query("nclass == 2")
```

```python
# 1반 수학 점수 평균
nclass1['math'].mean()
```

```python
# 2반 수학 점수 평균
nclass2['math'].mean()
```

```python
# 1반 영어 점수 평균
nclass1['english'].mean()
```

```python
# 2반 영어 점수 평균
nclass2['english'].mean()
```
